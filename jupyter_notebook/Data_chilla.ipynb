## Data_Chilla Graphs

### Countplot by Baba_aammar

```python
#import library
import pandas as pd

#import data from file
chilla = pd.read_csv("chilla_data1.csv")
print(chilla)

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)

p= sns.countplot(x="Gender", hue="Age",  data=chilla)
plt.show()
```

### Countplot with different variables

```python
import pandas as pd
chilla = pd.read_csv("chilla_data1.csv")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
p= sns.countplot(x="Location", hue="Time_of_class_(pm)",  data=chilla)
plt.show()
```

### Countplot with title

```python
import pandas as pd
chilla = pd.read_csv("chilla_data1.csv")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
p= sns.countplot(x="Location", hue="Time_of_class_(pm)",  data=chilla)
p.set_title("Data Chilla")
plt.show()
```
